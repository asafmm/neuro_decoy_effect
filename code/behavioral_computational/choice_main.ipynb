```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import read_files
import importlib
pd.options.display.float_format = "{:,.6f}".format
```

```python
all_choice_files = glob.glob(f'../../data/behavioral_experiment/*.csv')
print(f'Found {len(all_choice_files)} choice files')
```

```python
behavior_table, binary_ratios, trinary_ratios, raw_choices, subject_ratio_dfs = read_files.analyze_data_files(all_choice_files, verbose=True)
behavior_table = behavior_table[behavior_table.catch==0].copy()
```

```python
behavior_pretty_table = behavior_table[['binary_id', 'trinary_id', 'decoy_effect_A', 'A_ratio_trinary', 'A_ratio_binary', 'C_total', 'n_trinary', 'n_binary','decoy_direction']]
behavior_pretty_table = behavior_pretty_table.sort_values('decoy_effect_A')
```

```python
bootstrap_std = read_files.bootstrap_std(behavior_table, binary_ratios, trinary_ratios, B=10_000)
```

```python
behavior_pretty_table = behavior_pretty_table.merge(bootstrap_std, on=['binary_id'])
```
